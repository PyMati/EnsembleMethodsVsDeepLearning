```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import MinMaxScaler, LabelEncoder
```

```python
TRAINING_DATA_PATH = "./house_prices/train.csv"
TEST_DATA_PATH = "./house_prices/test.csv"
```

```python
training_data = pd.read_csv(TRAINING_DATA_PATH)
training_y = training_data['SalePrice']
training_data.drop('SalePrice', axis=1, inplace=True)
test_data = pd.read_csv(TEST_DATA_PATH)
```

```python
integer_type_columns = []
float_type_columns = []
object_type_columns = []
```

```python
for dtype, column_name in zip(training_data.dtypes, training_data.columns):
    if dtype == np.int64:
        integer_type_columns.append([column_name, dtype])
    elif dtype == np.float64:
        float_type_columns.append([column_name, dtype])
    else:
        object_type_columns.append([column_name, dtype])
```

```python
integer_type_columns
```

```python
len(object_type_columns) + len(integer_type_columns) + len(float_type_columns)
```

```python
def get_unique_values(column_name):
    return training_data[column_name].unique()
```

```python
unique_classes_values = list(map(lambda x: get_unique_values(x[0]), object_type_columns))
```

```python
classes_for_encoder = []
for unique_class_values in unique_classes_values:
    for value in unique_class_values:
        classes_for_encoder.append(value)
```

```python
len(classes_for_encoder)
```

```python
le = LabelEncoder()
le.fit(classes_for_encoder)
```

```python
le.classes_
```

```python
for class_column in object_type_columns:
    training_data[class_column[0]] = le.transform(training_data[class_column[0]])
    test_data[class_column[0]] = le.transform(test_data[class_column[0]])
```

```python
training_data.info()
```

```python
training_data.head()
```

```python
training_data = training_data.fillna(0)
test_data = test_data.fillna(0)
```

```python
training_data.isna().sum()
```

```python
scaler = MinMaxScaler()
scaler.fit(training_data)
```

```python
scaler.data_max_
```

```python
training_data = scaler.transform(training_data)
```

```python
test_data = scaler.transform(test_data)
```
